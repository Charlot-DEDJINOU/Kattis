```python
%matplotlib inline
import numpy as np
from math import exp
import matplotlib.pyplot as plt
plt.style.use('seaborn-whitegrid')
fig=plt.figure()
ax=plt.axes()
plt.title("Mon premier")
plt.axis([-11,11,0,1500])
x=range(-10,11)
y=[exp(abs(i)) for i in x]
plt.plot(x,y,color="green" ,linestyle="solid",label="green")
plt.plot(x,[exp(abs(i+1)) for i in x],color="yellow",linestyle="solid",label="yellow")
plt.plot(x,[exp(abs(i-1)) for i in x],color="red",linestyle="solid",label="red")
plt.plot(x,[exp(abs(i+2)) for i in x],color="blue",linestyle="solid",label="blue")
plt.plot(x,[exp(abs(i+3)) for i in x])
plt.legend(loc="lower right")
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
from math import sin
fig=plt.figure(figsize=(12,12))
x=range(-5,5)
for i in range(2) :
    fig.add_subplot(1,2,i+1)
    plt.style.use(plt.style.available[i])
    plt.scatter(x,[sin(j-i) for j in x])
    plt.text(s=plt.style.available[i],x=10,y=1,color="green")
```
