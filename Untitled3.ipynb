```python
%matplotlib inline
import matplotlib.pyplot as plt 
from math import sqrt
from numpy import log

f = lambda x :(1+sqrt(pow(x,2)+2))
d=range(-100,100)
y=[]
for x in d :
    y.append(f(x))
    
plot=plt.plot(d,y)
```
