```python
%matplotlib inline
import matplotlib.pyplot as plt
from math import exp
plt.style.use('seaborn-whitegrid')
fig=plt.figure()
axe=plt.axes()
plt.axis([-10,10,0,1500])
plt.title("Mes graphes")
x=range(-10,10)
y=[exp(abs(i)) for i in x]
plt.plot(x,y,color="green",linestyle="solid",label="green")
plt.plot(x,[exp(abs(i-1)) for i in x],color="yellow",linestyle="dotted",label="yellow")
plt.plot(x,[exp(abs(i-2)) for i in x],color="red",linestyle="dashed",label="red")
plt.plot(x,[exp(abs(i+1)) for i in x],color="blue",linestyle="solid",label="blue")
plt.legend(loc="lower right")
```
